```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/sample_data/placement.csv')
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.title('Scatter Plot')
plt.show()
```

```python
X=df.iloc[:,0:1]
Y=df.iloc[:,-1]
Y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)

```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
lr.predict(X_test.iloc[0].values.reshape(1, -1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_test,lr.predict(X_test),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.title('Scatter Plot')
plt.show()
```

```python
m=lr.coef_
```

```python
b=lr.intercept_
```

```python
class LinearRegression:
    def __init__(self):
      self.m=None
      self.b=None

    def fit(self,X_train,Y_train):

      num=0
      den=0
      for i in range(X_train.shape[0]):
        num=num+((X_train.iloc[i, 0]-X_train.mean())*(Y_train.iloc[i]-Y_train.mean()))
        den=den+((X_train.iloc[i, 0]-X_train.mean())*(X_train.iloc[i, 0]-X_train.mean()))


      self.m=num/den
      self.b=Y_train.mean()-(self.m*X_train.mean())
      print(self.m)
      print(self.b)

    def predict(self,X_test):
      print(X_test)
      return self.m+X_test+self.b
```

```python
lrMine=LinearRegression()
lrMine.fit(X_train,Y_train)
```

```python
lrMine.predict(X_test.iloc[0].values[0])
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
r2_score(Y_test,lrMine.predict(X_test))
```

```python
print("MAE",mean_absolute_error(Y_test,lrMine.predict(X_test)))
print("MSE",mean_squared_error(Y_test,lrMine.predict(X_test)))
```
