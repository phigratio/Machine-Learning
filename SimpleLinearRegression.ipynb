```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('/content/height_weight.csv')
df.head()
```

```python
plt.scatter(df['Weight_kg'],df['Height_cm'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Scatter Plot')
plt.show()
```

```python
df.corr()
```

```python
#Seaborn
import seaborn as sns
sns.pairplot(df)
```

```python
#Independent and dependent features
X=df[['Weight_kg']]
X
```

```python
np.array(X).shape
```

```python
X.head()
```

```python
Y=df['Height_cm']
Y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.26,random_state=2)
```

```python
X_train.shape
```

```python
## Standardization
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
```

```python
scaler.fit_transform(X_train)
```

```python
scaler.transform(X_test)
```

```python
## Apply Simple Linear Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
predicted_lr=lr.predict(X_train)
```

```python
plt.scatter(X_train,Y_train)
plt.plot(X_train,predicted_lr,color='red')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Linear Regression')
plt.show()
```

```python
y_pred=lr.predict(X_test)
```

```python
## Performance Metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mse=mean_squared_error(Y_test,y_pred)
mae=mean_absolute_error(Y_test,y_pred)
rmse=np.sqrt(mse)
print("MSE",mse)
print("MAE",mae)
print("RMSE",rmse)
```

```python
# Rsquare
from sklearn.metrics import r2_score
r2_score(Y_test,y_pred)
```
