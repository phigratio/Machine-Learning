```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ipl=pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
batsman=pd.read_csv('/content/batsman_season_record.csv')
batter=pd.read_csv('/content/batter.csv')
iris=pd.read_csv('/content/iris.csv')
tips=sns.load_dataset('tips')
```

```python
iris.head()
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='plasma')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

```python
sample=batter.sample(40)
```

```python
plt.figure(figsize=(12,10))
plt.scatter(sample['avg'],sample['strike_rate'],s=sample['runs'])
plt.xlabel('Average')
plt.axhline(130,color='green')
plt.axvline(20,color='red')
for i in range(sample.shape[0]):
  plt.text(sample['avg'].values[i],sample['strike_rate'].values[i],sample['batter'].values[i])
```

```python
#SubPlots
fig,ax=plt.subplots()
ax.scatter(sample['avg'],sample['strike_rate'])
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1)
ax[0].scatter(sample['avg'],sample['strike_rate'])
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike Rate')

ax[1].scatter(sample['avg'],sample['strike_rate'])
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Strike Rate')
```

```python
#3D scatter plots
batter

fig=plt.figure()
ax=plt.subplot(projection='3d')
ax.scatter(batter['avg'],batter['strike_rate'],batter['runs'])
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
ax.set_zlabel('Runs')
```

```python
#3D surface plots
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(10,10))
ax=plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
```
