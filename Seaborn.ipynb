```python
#Seaborn is simpler to use because of layer of abstraction
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Relational Plot
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(tips,x='total_bill',y='tip')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
#line
gap=px.data.gapminder()
temp_df=gap[gap['country']=='Bangladesh']
temp_df
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
temp_df=gap[gap['country'].isin(['Bangladesh','India','Pakistan'])]
temp_df
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',hue='country')
```

```python
temp_df=gap[gap['country'].isin(['Bangladesh','Germany','Brazil'])]
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',hue='country',style='continent')
```

```python
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```
